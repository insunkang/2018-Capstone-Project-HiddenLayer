# 아두이노 인식

```python
import serial
import MySQLdb

def isNumber(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

while True :
    
    ser = serial.Serial('/dev/cu.usbmodem1421',9600)
    ser.flushInput()
    a = 0
    
    for i in range(60):
        
        response = ser.readline()
        k = [response]
        
        if isNumber(k[0]) == True:
            a = a + int(k[0])
        elif isNumber(k[0]) == False:
            a = a
    #print a
    
    db = MySQLdb.connect("smurf1213.cafe24.com","smurf1213","1q2w3e4r!","smurf1213" )
    cursor = db.cursor()
    sql = "INSERT INTO Arduino(value) \
               VALUES ('%d')" % (a)
    cursor.execute(sql)
    db.commit()
    db.close()
```
